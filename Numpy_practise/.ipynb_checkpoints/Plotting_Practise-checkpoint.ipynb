```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(5)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,np.pi,100) #Gives us 10 points between 0 and Pi
```

```python
y = 2 * np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("$2*\cos(x)$") #Latex Code for looking good
```

```python
#Br Plot and Scatter Plot
vec = np.array([1,5,9,3,8,7])
num_day_x = np.arange(6)
```

```python
plt.bar(num_day_x,vec)
plt.xlabel("Day")
plt.ylabel("Rain Each day")
plt.title("My First Bar Plot")
plt.show()
```
